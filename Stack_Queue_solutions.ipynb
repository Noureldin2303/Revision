### Stack Solutions

#### Easy Level

1. **Create a stack and push the numbers 1 to 5 onto it. Print the stack.**
   ```python
   # Method 1
   stack = []
   for i in range(1, 6):
       stack.append(i)
   print(stack)  # Output: [1, 2, 3, 4, 5]

   # Method 2
   stack = [1, 2, 3, 4, 5]
   print(stack)  # Output: [1, 2, 3, 4, 5]
   ```

2. **Pop the top element from the stack and print the result.**
   ```python
   # Method 1
   stack = [1, 2, 3, 4, 5]
   top_element = stack.pop()
   print(top_element)  # Output: 5
   print(stack)        # Output: [1, 2, 3, 4]

   # Method 2
   stack = [1, 2, 3, 4, 5]
   print(stack.pop())  # Output: 5
   print(stack)        # Output: [1, 2, 3, 4]
   ```

3. **Check if the stack is empty and print the result.**
   ```python
   # Method 1
   stack = []
   is_empty = len(stack) == 0
   print(is_empty)  # Output: True

   # Method 2
   stack = []
   is_empty = not stack
   print(is_empty)  # Output: True
   ```

#### Intermediate Level

1. **Create a stack with the numbers 1 to 5. Use the `pop()` method to remove all elements and print each one.**
   ```python
   # Method 1
   stack = [1, 2, 3, 4, 5]
   while stack:
       print(stack.pop())  # Output: 5 4 3 2 1

   # Method 2
   stack = [1, 2, 3, 4, 5]
   for _ in range(len(stack)):
       print(stack.pop())  # Output: 5 4 3 2 1
   ```

2. **Create a stack with the numbers 1 to 5. Use the `peek()` method to get the top element without removing it, then push the number 6 and print the stack.**
   ```python
   # Method 1
   stack = [1, 2, 3, 4, 5]
   top_element = stack[-1]
   print(top_element)  # Output: 5
   stack.append(6)
   print(stack)        # Output: [1, 2, 3, 4, 5, 6]

   # Method 2
   stack = [1, 2, 3, 4, 5]
   print(stack[-1])  # Output: 5
   stack.append(6)
   print(stack)      # Output: [1, 2, 3, 4, 5, 6]
   ```

3. **Create a stack with the numbers 1 to 5. Use the `size()` method to get the number of elements, then pop two elements and print the new size.**
   ```python
   # Method 1
   stack = [1, 2, 3, 4, 5]
   size = len(stack)
   print(size)  # Output: 5
   stack.pop()
   stack.pop()
   new_size = len(stack)
   print(new_size)  # Output: 3

   # Method 2
   stack = [1, 2, 3, 4, 5]
   print(len(stack))  # Output: 5
   stack.pop()
   stack.pop()
   print(len(stack))  # Output: 3
   ```

#### Hard Level

1. **Create a stack with the numbers 1 to 5. Use a loop to pop all elements and store them in a list. Print the list.**
   ```python
   # Method 1
   stack = [1, 2, 3, 4, 5]
   popped_elements = []
   while stack:
       popped_elements.append(stack.pop())
   print(popped_elements)  # Output: [5, 4, 3, 2, 1]

   # Method 2
   stack = [1, 2, 3, 4, 5]
   popped_elements = [stack.pop() for _ in range(len(stack))]
   print(popped_elements)  # Output: [5, 4, 3, 2, 1]
   ```

2. **Create a stack with the numbers 1 to 5. Use the `peek()` method to get the top element, then push the top element multiplied by 2 back onto the stack. Print the stack.**
   ```python
   # Method 1
   stack = [1, 2, 3, 4, 5]
   top_element = stack[-1]
   stack.append(top_element * 2)
   print(stack)  # Output: [1, 2, 3, 4, 5, 10]

   # Method 2
   stack = [1, 2, 3, 4, 5]
   stack.append(stack[-1] * 2)
   print(stack)  # Output: [1, 2, 3, 4, 5, 10]
   ```

3. **Create a stack with the numbers 1 to 5. Use the `isEmpty()` method to check if the stack is empty, then pop all elements and check again. Print the results.**
   ```python
   # Method 1
   stack = [1, 2, 3, 4, 5]
   is_empty_before = len(stack) == 0
   while stack:
       stack.pop()
   is_empty_after = len(stack) == 0
   print(is_empty_before)  # Output: False
   print(is_empty_after)   # Output: True

   # Method 2
   stack = [1, 2, 3, 4, 5]
   is_empty_before = not stack
   stack.clear()
   is_empty_after = not stack
   print(is_empty_before)  # Output: False
   print(is_empty_after)   # Output: True
   ```

### Queue Solutions

#### Easy Level

1. **Create a queue and enqueue the numbers 1 to 5. Print the queue.**
   ```python
   # Method 1
   from collections import deque
   queue = deque()
   for i in range(1, 6):
       queue.append(i)
   print(queue)  # Output: deque([1, 2, 3, 4, 5])

   # Method 2
   queue = deque([1, 2, 3, 4, 5])
   print(queue)  # Output: deque([1, 2, 3, 4, 5])
   ```

2. **Dequeue the front element from the queue and print the result.**
   ```python
   # Method 1
   queue = deque([1, 2, 3, 4, 5])
   front_element = queue.popleft()
   print(front_element)  # Output: 1
   print(queue)          # Output: deque([2, 3, 4, 5])

   # Method 2
   queue = deque([1, 2, 3, 4, 5])
   print(queue.popleft())  # Output: 1
   print(queue)            # Output: deque([2, 3, 4, 5])
   ```

3. **Check if the queue is empty and print the result.**
   ```python
   # Method 1
   queue = deque()
   is_empty = len(queue) == 0
   print(is_empty)  # Output: True

   # Method 2
   queue = deque()
   is_empty = not queue
   print(is_empty)  # Output: True
   ```

#### Intermediate Level

1. **Create a queue with the numbers 1 to 5. Use the `dequeue()` method to remove all elements and print each one.**
   ```python
   # Method 1
   queue = deque([1, 2, 3, 4, 5])
   while queue:
       print(queue.popleft())  # Output: 1 2 3 4 5

   # Method 2
   queue = deque([1, 2, 3, 4, 5])
   for _ in range(len(queue)):
       print(queue.popleft())  # Output: 1 2 3 4 5
   ```

2. **Create a queue with the numbers 1 to 5. Use the `peek()` method to get the front element without removing it, then enqueue the number 6 and print the queue.**
   ```python
   # Method 1
   queue = deque([1, 2, 3, 4, 5])
   front_element = queue[0]
   print(front_element)  # Output: 1
   queue.append(6)
   print(queue)          # Output: deque([1, 2, 3, 4, 5, 6])

   # Method 2
   queue = deque([1, 2, 3, 4, 5])
   print(queue[0])  # Output: 1
   queue.append(6)
   print(queue)     # Output: deque([1, 2, 3, 4, 5, 6])
   ```

3. **Create a queue with the numbers 1 to 5. Use the `size()` method to get the number of elements, then dequeue two elements and print the new size.**
   ```python
   # Method 1
   queue = deque([1, 2, 3, 4, 5])
   size = len(queue)
   print(size)  # Output: 5
   queue.popleft()
   queue.popleft()
   new_size = len(queue)
   print(new_size)  # Output: 3

   # Method 2
   queue = deque([1, 2, 3, 4, 5])
   print(len(queue))  # Output: 5
   queue.popleft()
   queue.popleft()
   print(len(queue))  # Output: 3
   ```

#### Hard Level

1. **Create a queue with the numbers 1 to 5. Use a loop to dequeue all elements and store them in a list. Print the list.**
   ```python
   # Method 1
   queue = deque([1, 2, 3, 4, 5])
   dequeued_elements = []
   while queue:
       dequeued_elements.append(queue.popleft())
   print(dequeued_elements)  # Output: [1, 2, 3, 4, 5]

   # Method 2
   queue = deque([1, 2, 3, 4, 5])
   dequeued_elements = [queue.popleft() for _ in range(len(queue))]
   print(dequeued_elements)  # Output: [1, 2, 3, 4, 5]
   ```

2. **Create a queue with the numbers 1 to 5. Use the `peek()` method to get the front element, then enqueue the front element multiplied by 2 back onto the queue. Print the queue.**

   ```python
   # Method 2
   queue = deque([1, 2, 3, 4, 5])
   queue.append(queue[0] * 2)
   print(queue)  # Output: deque([1, 2, 3, 4, 5, 2])
   ```

3. **Create a queue with the numbers 1 to 5. Use the `isEmpty()` method to check if the queue is empty, then dequeue all elements and check again. Print the results.**

   ```python
   # Method 1
   queue = deque([1, 2, 3, 4, 5])
   is_empty_before = len(queue) == 0
   ```
