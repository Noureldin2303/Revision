#### Hard Level

1. **Create a binary tree with the root node value 10. Insert nodes with values 5, 15, 3, 7, 12, and 18. Write a function to find the height of the tree.**
   ```python
   def height(node):
       if node is None:
           return 0
       else:
           left_height = height(node.left)
           right_height = height(node.right)
           return max(left_height, right_height) + 1

   # Method 1
   root = Node(10)
   root.left = Node(5)
   root.right = Node(15)
   root.left.left = Node(3)
   root.left.right = Node(7)
   root.right.left = Node(12)
   root.right.right = Node(18)
   print(height(root))  # Output: 3

   # Method 2
   root = Node(10)
   insert(root, 5)
   insert(root, 15)
   insert(root, 3)
   insert(root, 7)
   insert(root, 12)
   insert(root, 18)
   print(height(root))  # Output: 3
   ```

   ```python
   def count_nodes(node):
       if node is None:
           return 0
       else:
           return 1 + count_nodes(node.left) + count_nodes(node.right)

   # Method 1

   root.left = Node(5)
   root.right = Node(15)
   root.left.left = Node(3)
   root.left.right = Node(7)

   # Method 2
   root = Node(10)
   insert(root, 5)
   insert(root, 15)
   insert(root, 3)
   insert(root, 7)
   insert(root, 12)
   insert(root, 18)
   print(count_nodes(root))  # Output: 7
   ```

3. **Create a binary tree with the root node value 10. Insert nodes with values 5, 15, 3, 7, 12, and 18. Write a function to check if the tree is a binary search tree (BST).**
   ```python
   def is_bst(node, left=float('-inf'), right=float('inf')):
       if node is None:
           return True
       if not (left < node.val < right):
           return False
       return is_bst(node.left, left, node.val) and is_bst(node.right, node.val, right)

   # Method 1
   root = Node(10)
   root.left = Node(5)
   root.right = Node(15)
   root.left.left = Node(3)
   root.left.right = Node(7)
   root.right.left = Node(12)
   root.right.right = Node(18)
   print(is_bst(root))  # Output: True

   # Method 2
   root = Node(10)
   insert(root, 5)
   insert(root, 15)
   insert(root, 3)
   insert(root, 7)
   insert(root, 12)
   insert(root, 18)
   print(is_bst(root))  # Output: True
   ```
