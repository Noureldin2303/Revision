### Stacks

A stack is a LIFO (Last In First Out) data structure. The main operations are `push` (to add an element to the top) and `pop` (to remove the top element).

#### Stack Methods

1. **push()**: Adds an element to the top of the stack.
   ```python
   stack = []
   stack.append(1)
   stack.append(2)
   print(stack)  # Output: [1, 2]
   ```

2. **pop()**: Removes and returns the top element of the stack.
   ```python
   stack = [1, 2, 3]
   top_element = stack.pop()
   print(top_element)  # Output: 3
   print(stack)        # Output: [1, 2]
   ```

3. **peek()**: Returns the top element without removing it.
   ```python
   stack = [1, 2, 3]
   top_element = stack[-1]
   print(top_element)  # Output: 3
   ```

4. **isEmpty()**: Checks if the stack is empty.
   ```python
   stack = []
   is_empty = len(stack) == 0
   print(is_empty)  # Output: True
   ```

5. **size()**: Returns the number of elements in the stack.
   ```python
   stack = [1, 2, 3]
   size = len(stack)
   print(size)  # Output: 3
   ```

### Queues

A queue is a FIFO (First In First Out) data structure. The main operations are `enqueue` (to add an element to the end) and `dequeue` (to remove the front element).

#### Queue Methods

1. **enqueue()**: Adds an element to the end of the queue.
   ```python
   from collections import deque
   queue = deque()
   queue.append(1)
   queue.append(2)
   print(queue)  # Output: deque([1, 2])
   ```

2. **dequeue()**: Removes and returns the front element of the queue.
   ```python
   queue = deque([1, 2, 3])
   front_element = queue.popleft()
   print(front_element)  # Output: 1
   print(queue)          # Output: deque([2, 3])
   ```

3. **peek()**: Returns the front element without removing it.
   ```python
   queue = deque([1, 2, 3])
   front_element = queue[0]
   print(front_element)  # Output: 1
   ```

4. **isEmpty()**: Checks if the queue is empty.
   ```python
   queue = deque()
   is_empty = len(queue) == 0
   print(is_empty)  # Output: True
   ```

5. **size()**: Returns the number of elements in the queue.
   ```python
   queue = deque([1, 2, 3])
   size = len(queue)
   print(size)  # Output: 3
   ```
